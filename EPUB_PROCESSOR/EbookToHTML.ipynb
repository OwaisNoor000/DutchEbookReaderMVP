```python
import os
from ebooklib import epub
from bs4 import BeautifulSoup
from weasyprint import HTML
from PIL import Image
```

```python
epub_file = 'C:/Users/mowai/Documents/Projects/EPUB reader/Modules/EPUB/Dolfje_Weerwolfje.epub'  # Path to your EPUB file
```

```python
# Function to extract HTML content from an EPUB file

# Read the EPUB file
book = epub.read_epub(epub_file)

# Prepare the output folder
output_dir = "extracted_html"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
```

```python

# Iterate through the EPUB items and extract HTML
html_content = []
for item in book.get_items():
    if item.get_type() == 9:
        # Parse the HTML content with BeautifulSoup for easier handling
        soup = BeautifulSoup(item.get_body_content(), 'html.parser')
        
        # You can clean up or modify the HTML here if needed
        
        # Save the extracted HTML content to a file
        html_filename = os.path.join(output_dir, item.get_id() + ".html")
        with open(html_filename, 'w', encoding='utf-8') as f:
            f.write(str(soup))

        html_content.append(str(soup))

print(html_content)
```

```python
html = HTML(string=html_content[0])

# Render the HTML as a PNG image (you can adjust size/format as needed)
image = html.write_png()

# Save the image as a temporary file
image_path = "page1.png"
with open(image_path, 'wb') as f:
    f.write(image)
```
