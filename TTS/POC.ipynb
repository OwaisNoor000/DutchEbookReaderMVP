```python
characters = alignment.characters
start_times = alignment.character_start_times_seconds
end_times = alignment.character_end_times_seconds
```

```python
# get stop indexes
stop_indexes=[]

for i,char in enumerate(characters):
    if "." in char:
        stop_indexes.append(i)
```

```python
import subprocess

def split_mp3(mp3_file, end_times):
    start_time = 0
    for i, end_time in enumerate(end_times):
        output_file = f"chunk_{i}.mp3"
        duration = end_time - start_time
        command = [
            "ffmpeg", "-i", mp3_file, "-ss", str(start_time),
            "-t", str(duration), "-c", "copy", output_file, "-y"
        ]
        subprocess.run(command, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
        start_time = end_time  # Move to next chunk
        print(f"Saved: {output_file}")

```

```python
# Example usage
mp3_filename = "output_audio.mp3"
split_mp3(mp3_filename, sentence_end_times)
```
