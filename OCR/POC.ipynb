## Getting Bounding boxes and Text mappings

```python
data = pytesseract.image_to_data(image,output_type=pytesseract.Output.DICT)
```

```python
data["text"]
```

```python
for i in range(len(data["text"])):
    if data["text"][i].strip():  # Ignore empty words
        x, y, w, h = data["left"][i], data["top"][i], data["width"][i], data["height"][i]
        print(f"Word: '{data['text'][i]}' | Position: (x={x}, y={y}, w={w}, h={h})")
```

```python
# Draw bounding boxes on the image
for i in range(len(data["text"])):
    if data["text"][i].strip():
        x, y, w, h = data["left"][i], data["top"][i], data["width"][i], data["height"][i]
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
        cv2.putText(image, data["text"][i], (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1)

# Show the image
cv2.imshow("Word Bounding Boxes", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
